# Series from lists


```python
# string

country = ['India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
# integers

runs = [13,24,56,78,100]
runs_ser = pd.Series(runs)
runs_ser
```

```python
# custom index

marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']
pd.Series(marks,index=subjects)
```

```python
# setting a name

marks = pd.Series(marks,index=subjects,name='Pal ke marks')
marks
```

# Series from dict


```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks,name='Pal ke marks')
marks_series
```
